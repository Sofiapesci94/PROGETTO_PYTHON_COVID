```python
# 1. Identifichiamo i delimitatori per i colori
delimiters = ['-']

# 2. Splitta le stringhe
# prima sostituiamo tutti i delimitatori (e.g., ',')
df_uomo['colori_split'] = df_uomo['colori'].str.replace(r'[-]', ',', regex=True)
# Then, split by the consistent delimiter
df_uomo['colori_split'] = df_uomo['colori_split'].str.split(',')

# 3. la lista la trasformiamo in un dataframe
df_colors = df_uomo.explode('colori_split')

# 4. pulizia
df_colors['colori_split'] = df_colors['colori_split'].str.strip().str.lower()

# 5. Analiza i colori
color_counts = df_colors['colori_split'].value_counts().nlargest(25) 
print(color_counts)
```
