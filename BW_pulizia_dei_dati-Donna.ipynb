#### Verifichiamo che le condizioni sono state soddisfatte

```python

def modifica_dataframe_multiplo(df):
    """
    Modifica il DataFrame dove la colonna 'sport' contiene una delle parole chiave specificate.

    Args:
        df (pandas.DataFrame): Il DataFrame da modificare.

    Returns:
        pandas.DataFrame: Il DataFrame modificato.
    """
    parole_chiave = [
        'sonoma 7 gtx', 'sonoma 8 gtx', 'venture 10', '1000 13 gtx', 'trabuco 13', 'nimbus 27',
        '1000 13 tr', 'trac nitro', 'pulse 16', '2000 13 tr', '1000 13', 'venture 10 wp', 'sonoma 8',
        'kayano 31', 'cumulus 27 lite', 'trabuco 12 gtx', 'excite 10', 'mission', 'nimbus 27 tr',
        'cumulus 27', '2000 12 gtx', 'contend 9', 'venture 9', 'nimbus 27 lite', '2000 13',
        'dedicate 8', 'resolution x clay', 'excite 2', 'dedicate 8 clay', 'on', 'rocket 11',
        'x efekt', 'nimbus 26', 'tactic 12', 'task 4', 'task mt 4', 'resolution', 'challenger 14',
        'cumulus 26', 'court hunter 3', 'cumulus 25', 'cumulus 26 gtx', 'challenger 14 clay',
        'resolution x', 'pulse 15', 'pro nitro dylan exhibit', 'game 9 clay', 'cell nova fs ultra',
        'pro nitro basketball', 'game 9', '1130'
    ]

    # Crea una condizione che verifica se la colonna 'sport' contiene una qualsiasi delle parole chiave
    condizione = df['sport'].str.contains('|'.join(fr'\b{parola}\b' for parola in parole_chiave), case=False, na=False, regex=True)

    # Applica le modifiche solo a queste righe
    df.loc[condizione, 'modello'] = df.loc[condizione, 'modello'].astype(str) + ' ' + df.loc[condizione, 'sport'].astype(str)

    # Splitta la colonna 'colori'
    colori_splitted = df.loc[condizione, 'colori'].str.split('-', n=1, expand=True)

    # La prima parte dello split viene passata alla colonna 'sport'
    df.loc[condizione, 'sport'] = colori_splitted[0]

    # La seconda parte dello split (se esiste) viene mantenuta in 'colori'
    # Se non c'è '-', la colonna 'colori' diventa NaN
    df.loc[condizione, 'colori'] = colori_splitted[1].fillna(pd.NA)

    return df

# Applica la funzione al DataFrame
df = modifica_dataframe_multiplo(df.copy())
```
