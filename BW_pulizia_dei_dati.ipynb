```python
df_scarpe= df[df['sport'] == 'phoenix fire']
df_scarpe
```

```python
def inverti_modello_sport_phoenix(df):
    """
    Inverte il contenuto delle colonne 'modello' e 'sport'
    solo dove la colonna 'sport' contiene 'phoenix fire'.

    Args:
        df (pandas.DataFrame): Il DataFrame da modificare.

    Returns:
        pandas.DataFrame: Il DataFrame modificato.
    """
    # Identifica le righe dove la colonna 'sport' contiene 'phoenix fire'
    condizione_phoenix = df['sport'].str.contains('phoenix fire', case=False, na=False)

    # Inverti i valori delle colonne 'modello' e 'sport' per le righe selezionate
    # Utilizziamo una variabile temporanea per evitare sovrascritture
    temp_modello = df.loc[condizione_phoenix, 'modello'].copy()
    df.loc[condizione_phoenix, 'modello'] = df.loc[condizione_phoenix, 'sport']
    df.loc[condizione_phoenix, 'sport'] = temp_modello

    return df


# Applica la funzione al tuo DataFrame e riassegna il risultato a 'df'
df = inverti_modello_sport_phoenix(df)
```

```python
def modifica_dataframe(df):

    # Identifica le righe dove la colonna 'sport' contiene la parola sbagliata
    condizione = df['sport'].str.contains('game 9 clay', case=False, na=False)

    # Applica le modifiche solo a queste righe
    df.loc[condizione, 'modello'] = df.loc[condizione, 'modello'] + ' ' + df.loc[condizione, 'sport']

    # Splitta la colonna 'colori'
    colori_splitted = df.loc[condizione, 'colori'].str.split('-', n=1, expand=True)

    # La prima parte dello split viene passata alla colonna 'sport'
    df.loc[condizione, 'sport'] = colori_splitted[0]

    # La seconda parte dello split (se esiste) viene mantenuta in 'colori'
    # Se non c'è '-', la colonna 'colori' diventa NaN
    df.loc[condizione, 'colori'] = colori_splitted[1].fillna(pd.NA)

    return df


# Applica la funzione al DataFrame
df = modifica_dataframe(df)
```
