```python
def modifica_dataframe(df):
    """
    Modifica il DataFrame solo dove la colonna 'sport' è 'flat dark gray'.

    Args:
        df (pandas.DataFrame): Il DataFrame da modificare.

    Returns:
        pandas.DataFrame: Il DataFrame modificato.
    """
    # Identifica le righe dove la colonna 'sport' è esattamente 'flat dark gray'
    condizione = df['sport'] == 'flat dark gray'

    # Unisci il contenuto della colonna 'sport' alla colonna 'colori'
    df.loc[condizione, 'colori'] = df.loc[condizione, 'colori'].astype(str) + ' ' + df.loc[condizione, 'sport'].astype(str)

    # Sposta il contenuto della colonna 'modello' in una nuova colonna temporanea
    df.loc[condizione, 'modello_temp'] = df.loc[condizione, 'modello']

    # Metti "Unknown" nella colonna 'modello'
    df.loc[condizione, 'modello'] = 'Unknown'

    # Rimuovi la colonna temporanea
    df = df.drop(columns=['modello_temp'], errors='ignore') # errors='ignore' se la colonna non esiste

    return df


# Applica la funzione al DataFrame
df = modifica_dataframe(df)
```

```python
# Filtrare le righe dove 'sport' è 'punte'
df_punte = df[df['sport'] == 'punte'].copy()

# Eliminare le colonne 'modello', 'sport', 'colori' e tutto ciò che viene dopo
df_punte = df_punte.drop(columns=['modello', 'sport', 'colori'])

# Splittare la colonna 'nome_prodotto' per riempire le nuove colonne
df_punte[['modello', 'sport', 'colori']] = df_punte['nome_prodotto'].str.split(' - ', n=2, expand=True)

# Verifica del nuovo DataFrame
display(df_punte)
```

```python
# Sostituire la colonna 'modello' con 'sun steam'
df_punte['modello'] = 'sun steam'

# Sostituire la colonna 'sport' con 'scarpe da corsa su strada'
df_punte['sport'] = 'scarpe da corsa su strada'

# Funzione per mantenere solo i colori validi nella colonna 'colori'
colori_validi = {'black', 'white', 'red', 'blue', 'green', 'yellow', 'silver', 'gold', 'gray', 'brown', 'pink', 'purple', 'orange'}  

def filtra_colori(testo):
    if pd.isna(testo):
        return testo  # Mantiene i valori NaN
    parole = testo.lower().split()  # Divide le parole
    colori_filtrati = [parola for parola in parole if parola in colori_validi]  # Mantiene solo i colori validi
    return ' '.join(colori_filtrati)  # Ricompone la stringa con solo i colori validi

# Applicare la funzione alla colonna 'colori'
df_punte['colori'] = df_punte['colori'].apply(filtra_colori)
```

```python
# Rimuovere le righe dove sport è 'punte' nel DataFrame originale
df = df[df['sport'] != 'punte']

# Unire df_punte al DataFrame originale
df = pd.concat([df, df_punte], ignore_index=True)
```
